```python
import numpy as np

timings = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
def checkins_hourly(input_data):

    #print(input_data)
    hourly_checkins = np.zeros(24*7)

    total_len = int(len(input_data.split(','))/2)

    for val in range(total_len):

        day_week = input_data.split(',')[val]
        day_hour = input_data.split(',')[val+total_len]
        index = int(day_week) * 24 + int(day_hour)

        hourly_checkins[index] += 1
    return hourly_checkins
#print(hourly_checkins)
```

```python
result_df['hourly_checkins'] = result_df['combined'].apply(lambda x: checkins_hourly(x))
```
